```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://github.com/LuisM78/Appliances-energy-prediction-data/raw/master/energydata_complete.csv"
energy_df = pd.read_csv(url)

# Convert 'date' column to datetime format
energy_df['date'] = pd.to_datetime(energy_df['date'])

# Display first few rows
print(energy_df.head())
```

```python
# Plot energy consumption over time
plt.figure(figsize=(15,5))
plt.plot(energy_df['date'], energy_df['Appliances'], label="Appliances Energy Consumption")
plt.xlabel('Time')
plt.ylabel('Energy Consumption (Wh)')
plt.title('Appliances Energy Consumption Over Time')
plt.legend()
plt.show()
```

```python
one_week_df = energy_df[(energy_df['date'] >= "2016-01-10") & (energy_df['date'] < "2016-01-17")]

plt.figure(figsize=(15,5))
plt.plot(one_week_df['date'], one_week_df['Appliances'])
plt.xlabel('Time')
plt.ylabel('Energy Consumption (Wh)')
plt.title('Appliances Energy Consumption Over One Week')
plt.show()
```

```python
energy_df['Hour'] = energy_df['date'].dt.hour
energy_df['Day'] = energy_df['date'].dt.date
heatmap_data = energy_df.pivot_table(index='Hour', columns='Day', values='Appliances', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=False)
plt.title('Heatmap of Hourly Energy Consumption Over a Week')
plt.show()
```

```python
plt.hist(energy_df['Appliances'], bins=30, edgecolor='black')
plt.xlabel('Energy Consumption (Wh)')
plt.ylabel('Frequency')
plt.title('Histogram of Energy Consumption')
plt.show()
```

```python
energy_df['NSM'] = energy_df['date'].dt.hour * 3600 + energy_df['date'].dt.minute * 60

plt.scatter(energy_df['NSM'], energy_df['Appliances'], alpha=0.5)
plt.xlabel('Seconds from Midnight')
plt.ylabel('Energy Consumption (Wh)')
plt.title('Energy Consumption vs NSM')
plt.show()
```

```python
plt.scatter(energy_df['Press_mm_hg'], energy_df['Appliances'], alpha=0.5)
plt.xlabel('Pressure (mm Hg)')
plt.ylabel('Energy Consumption (Wh)')
plt.title('Energy Consumption vs Pressure')
plt.show()
```
